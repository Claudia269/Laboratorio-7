```python
# Crear un array de números del 1 al 10
# mostrar la suma, la media, y el valor máximo.
```

```python
array=np.arange(1,11)
print(array)
suma = np.sum(array)
print("Suma:", suma)
media = np.mean(array)
print("Media:", media)
maximo = np.max(array)
print("Máximo:", maximo)

```

```python
# Crear un array de 3x4 lleno de ceros
#y luego cambiar su forma a 2x6
array_ceros = np.zeros((3, 4))
print("Array original:\n", array_ceros)
array_modificado = array_ceros.reshape(2, 6)
print("\nArray modificado:\n", array_modificado)
```

```python
matriz_aleatoria = np.random.rand(3, 3)
print("Matriz aleatoria:\n", matriz_aleatoria)
matriz_transpuesta = matriz_aleatoria.T
print("\nMatriz transpuesta:\n", matriz_transpuesta)

```
